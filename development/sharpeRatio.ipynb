```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import timedelta
import random as rng
from pathlib import Path
```

```python
df = pd.read_pickle('template.pkl')
# spyDf = pd.read_pickle('shortSPY.pkl')
df = df[df.index - pd.to_datetime('1999-06-01') > timedelta(0)]
# df['logGrowth'] = np.log(df['total'])
# df['logGrowth'].plot()
# df.plot()
df['randomTotal1'] = 100
df['randomTotal2'] = 100

for i in range(1,len(df)):
    df.iat[i,2] = df.iat[i-1,2] * rng.uniform(0.990, 1.014)
    df.iat[i,3] = df.iat[i-1,3] * rng.uniform(0.995, 1.009)
df['randomTotal1'].plot()
df['randomTotal2'].plot()
```

```python
def getSharpeRatio(df):
    df.plot()
    plt.show()
    dailyRet = df.pct_change()
    # excessRet = dailyRet - nasdaqDif
    excessRet = dailyRet - 0.04/52
    sharpeRatio = np.sqrt(52)*np.mean(excessRet) / np.std(excessRet)
    return sharpeRatio
```
