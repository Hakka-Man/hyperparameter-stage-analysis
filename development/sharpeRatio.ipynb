```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import timedelta
```

```python
df = pd.read_pickle('transactionExample.pkl')
df = df[df.index - pd.to_datetime('1999-06-01') > timedelta(0)]
df['logGrowth'] = np.log(df['total'])
df['logGrowth'].plot()
```

```python
dailyRet = df.loc[:, 'total'].pct_change()
nasdaqDf = pd.read_pickle("stockData/Spy.pkl")
nasdaqDif = nasdaqDf.loc[:, 'close'].pct_change()
excessRet = dailyRet - nasdaqDif
sharpeRatio = np.sqrt(52)*np.mean(excessRet) / np.std(excessRet)
sharpeRatio
```
