```python
import pandas as pd
pd.options.mode.chained_assignment = None  # default='warn'
import numpy as np
from datetime import date, timedelta
import yahoo_fin.stock_info as yf
from stage import fullPrint
```

```python
spDf = yf.get_data('^GSPC', start_date="1995-01-02",end_date="2022-01-17", index_as_date = True, interval="1wk")
```

```python
def calculateSlopeOfTicker(df,index):
    if index < 31:
        return 0
    else:
        return df.iat[index,7]/df.iat[index-1,7]
weights = np.arange(1,31)
sumWeights = np.sum(weights)
spDf['WMA30'] = spDf['close'].rolling(window=30).apply(lambda x: np.sum(weights*x)/sumWeights)
spDf['WMA30Slope'] = spDf.apply(lambda x: calculateSlopeOfTicker(spDf,spDf.index.get_loc(x.name)),axis=1)

spDf.to_pickle('stockData/Spy.pkl')
fullPrint(spDf['WMA30Slope'])
```
