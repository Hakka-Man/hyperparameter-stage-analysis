```python
## IMPORTS
import pandas as pd
import pandas_datareader.data as pdr
from pandas import Series, DataFrame
pd.options.mode.chained_assignment = None  # default='warn'
import numpy as np
import datetime
from datetime import date, timedelta
import yfinance as yf
yf.pdr_override()
```

```python
R3000TXT = open("stockData/russell3000Symbols.txt", "r")
R3000List = R3000TXT.read()
dfR3000Tickers = R3000List.split(",")
R3000TXT.close()
```

```python
dfSPYTickers = pd.read_csv('stockData/S&P500Symbols.csv')
```

```python
today = date.today().strftime('%Y-%m-%d')
sp = "^GSPC"
spdf = pdr.get_data_yahoo(sp, start="1995-01-01",end=today)
sortLogic = {
    'Close': 'last',
    'Volume': 'sum'
}
spDfSorted = spdf.resample("W-FRI").agg(sortLogic)
    
```

```python
for ticker in dfR3000Tickers:
    try:
        df = pdr.get_data_yahoo(ticker, start="1995-01-01",end=today)
        dfSorted = df.resample("W-FRI").agg(sortLogic)
        dfSorted.to_csv('stockData/R3000/'+ticker+'.csv')
    except:
        print("stock not found")
```

```python
for ticker in dfSPYTickers:
    try:
        df = pdr.get_data_yahoo(ticker, start="1995-01-01",end=today)
        dfSorted = df.resample("W-FRI").agg(sortLogic)
        dfSorted.to_csv('stockData/S&P500/'+ticker+'.csv')
    except:
        print("stock not found")
```
